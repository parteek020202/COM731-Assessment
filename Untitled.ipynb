```python
import pandas as pd
df_main = pd.read_csv("lung_cancer_data.csv", encoding="UTF-8")
print("Data Loaded Successfully!")
df_main
```

```python
ethnicity = "Hispanic"
df_filtered = df_main.loc[(df_main["Ethnicity"]==ethnicity) & (df_main["Survival_Months"]>100), ["Ethnicity","Survival_Months","Treatment"]]
top_treatments = (df_filtered["Treatment"].value_counts().head(3))
top_treatments
```
